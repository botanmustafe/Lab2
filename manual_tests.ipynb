```python
# manualtest circle.py

from circle import Circle

# create two circles

circle1 = Circle(x=0, y=0, radius=1)
circle2 = Circle(x=1, y=1, radius=1)

# test comparisons and methods

print(circle1 == circle2)
circle1.translate(5, 3)
print(circle1)

# test error handling

try:
    circle1.translate("THREE", 5)
except TypeError as e:
    print(":", e)

# test comparisons (area)

circle3 = Circle(radius=3)
print(circle3 > circle1)
print(circle3.is_unit_circle())




```

```python
# manualtest rectangle.py

from rectangle import Rectangle

# create two rectangles
rect1 = Rectangle(x=0, y=0, width=2, height=3)
rect2 = Rectangle(x=1, y=1, width=3, height=2)

# test equality (same size but different position)
print(rect1 == rect2)

# move rect1
rect1.translate(4, -1)
print(rect1)

# test is_square
rect3 = Rectangle(width=4, height=4)
print(rect3.is_square())

# test comparison by area
rect_small = Rectangle(width=1, height=2)
rect_big = Rectangle(width=5, height=5)
print(rect_small < rect_big)

# test error handling
try:
    rect1.translate("hello", 5)
except TypeError as e:
    print("Error:", e)
```
